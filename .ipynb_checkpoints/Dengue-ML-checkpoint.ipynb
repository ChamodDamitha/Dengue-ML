```python
# sns.set(style="white")
# df = X_train.loc[:,['reanalysis_min_air_temp_k','reanalysis_tdtr_k','station_diur_temp_rng_c']]
# g = sns.PairGrid(df, diag_sharey=False)
# # g.map_lower(sns.kdeplot, cmap="Blues_d")
# g.map_upper(plt.scatter)
# g.map_diag(sns.kdeplot, lw=3)

# Selecting highly correlated features with target var
correlation = all_data.corr(method='pearson')
selected_columns = correlation.nlargest(13, 'total_cases').index
selected_columns = list(selected_columns)
selected_columns.remove('total_cases')

y_columns = ['city', 'year', 'weekofyear']
for c in y_columns:
    if c not in selected_columns: selected_columns.append(c)

print(selected_columns)
```

```python
from sklearn.preprocessing import Imputer
    
def formatX(x) :
    # Add new columns
#     avg_ndvi = x[ndvi_columns].mean(axis=1)
#     x['ndvi_avg'] = avg_ndvi
        
    # Drop irrelevant columns
#     x = x.drop(['week_start_date', 
#                 'station_max_temp_c',
#                 'reanalysis_sat_precip_amt_mm', 'reanalysis_precip_amt_kg_per_m2'] 
#                + ndvi_columns + satelite_columns, 
#                axis=1)
    x = x.drop(['week_start_date'], 
               axis=1)
    
    # select columns     
    x = x[selected_columns]

    # Processing categoricals
    #One hot encoding categoricals
    x = pd.get_dummies(x)
    #without categoricals
    x = x.select_dtypes(exclude=['object'])

    # Identify columns with null values 
    print("Total Colums : " + str(len(x.columns)))
    # make new columns indicating what will be imputed
#     cols_with_missing = [col for col in x.columns if x[col].isnull().any()]
#     not_null_columns = [col for col in list(x.columns) if col not in cols_with_missing]
#     print("Columns with null values : " + str(len(cols_with_missing)))
#     print("Columns without null values : " + str(len(not_null_columns)))

    # Impute null values
    new_X = x.copy()
    my_imputer = Imputer()
    x = pd.DataFrame(my_imputer.fit_transform(new_X))
#     x = pd.concat([x, temp_X], axis=1)

    # cols_with_missing = [col for col in X.columns if X[col].isnull().any()]
    # print("Columns with null values : " + str(len(cols_with_missing)))

    # len(temp_X.columns)
    return x
# X_train
X = formatX(X_train)
X.head()
```

```python
from sklearn.model_selection import train_test_split

# split data into training and validation data, for both predictors and target
# The split is based on a random number generator. Supplying a numeric value to
# the random_state argument guarantees we get the same split every time we
# run this script.
train_X, val_X, train_y, val_y = train_test_split(X, y_train, test_size=0.1, random_state=0, shuffle=False)

val_X.head()
```

```python
# Build and Train model
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor

y = y_train['total_cases'] 

y2 = train_y['total_cases'] 

# model = RandomForestRegressor(n_estimators=200, max_features='auto', max_depth=7, min_samples_leaf=0.005)
# forest_model.fit(train_X, y)

# model = Lasso()
# model = ElasticNet()
# model = GradientBoostingRegressor(random_state=0, n_estimators=400)
# model = DecisionTreeRegressor(random_state=0)
# model = KNeighborsRegressor(random_state=0)

model.fit(train_X, y2)
# X.dtypes
# y.head()

# X.to_csv('X.csv')

from sklearn.metrics import mean_absolute_error
pred_val = model.predict(val_X)
print("MAE : " + str(mean_absolute_error(val_y['total_cases'], pred_val)))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

pipelines = []
pipelines.append(('ScaledLR', Pipeline([('Scaler', StandardScaler()),('LR',LinearRegression())])))
pipelines.append(('ScaledLASSO', Pipeline([('Scaler', StandardScaler()),('LASSO', Lasso())])))
pipelines.append(('ScaledEN', Pipeline([('Scaler', StandardScaler()),('EN', ElasticNet())])))
pipelines.append(('ScaledKNN', Pipeline([('Scaler', StandardScaler()),('KNN', KNeighborsRegressor())])))
pipelines.append(('ScaledCART', Pipeline([('Scaler', StandardScaler()),('CART', DecisionTreeRegressor())])))
pipelines.append(('ScaledGBM', Pipeline([('Scaler', StandardScaler()),('GBM', GradientBoostingRegressor())])))
pipelines.append(('ScaledRF', Pipeline([('Scaler', StandardScaler()),('RF', RandomForestRegressor(n_estimators=200, max_features='auto', max_depth=7, min_samples_leaf=0.005))])))

results = []
names = []
for name, model in pipelines:
    kfold = KFold(n_splits=10, random_state=21)
    cv_results = cross_val_score(model, X, y, cv=kfold, scoring='neg_mean_squared_error')
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)

```

```python
X_test = pd.read_csv(test_feature_filename)

X_test.head()

X_p = formatX(X_test)

X_p.head()

pred = pd.DataFrame(model.predict(X_p))
pred = pred.round().astype(int)
pred.columns = ['total_cases']

output = pd.concat([X_test[['city', 'year', 'weekofyear']], pred], axis=1)
output.head()
output.to_csv('submission.csv',  index = False)
```
